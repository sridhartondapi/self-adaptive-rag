```python
import boto3
import json
from langchain_aws import ChatBedrock
import requests

bedrock = boto3.client(service_name="bedrock-runtime",
                      region_name='us-east-1')

MODEL_ID = "us.anthropic.claude-3-7-sonnet-20250219-v1:0"  # inference profile ID
                      

def bedrock_connection():
    try:
        return ChatBedrock(
            client = bedrock,
            model_id = MODEL_ID,
            model_kwargs = {
                "temperature": 0.2,
                "max_tokens": 512
            }
        )
    except Exception as e:
        print (f"Error in connecting to the bedrock model")
```

```python
try:
    llm = bedrock_connection()
    print ("SUCCESS")
    #resp = llm.invoke("Hello from my laptop")
    #print(resp.content)
except Exception as e:
    print ("unable to connect to the bedrock")
```
